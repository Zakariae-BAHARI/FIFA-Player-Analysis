```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# the cleaned moroccan players data
moroccan_players = pd.read_csv('moroccan_players.csv')
```

```python
#basic information about Moroccan players
print(f"\nBasic Information about {'Moroccan'} National Team Players:")
print(moroccan_players[['Name', 'Age', 'BestPosition', 'Overall', 'Potential', 'ValueEUR', 'WageEUR']])
```

#  Age and Performance :

```python
# Age distribution
plt.figure(figsize=(10, 6))
sns.histplot(moroccan_players['Age'], bins=20, kde=True, color='skyblue')
plt.title('Age Distribution of Players')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
age_statistics = moroccan_players['Age'].describe()
print(age_statistics)
```

# Positional Analysis

```python
# Moroccan Players Distribution by Position 

position_counts = moroccan_players['BestPosition'].value_counts()

plt.figure(figsize=(12, 6))
sns.barplot(x=position_counts.index, y=position_counts.head(10))
plt.title('number of players in each position in the national team')
plt.xlabel('Player Position')
plt.ylabel('Number of Players')
plt.xticks(rotation=45)
plt.show()

# Print the count statistics about  player positions in the moroccan national team
print(f'the number of players in each position in the national team')
print(position_counts)
```

#  Financial Analysis

```python
# Financial Analysis for Moroccan National Team Players
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.histplot(moroccan_players['ValueEUR'], bins=20, kde=True, color='skyblue')
plt.title(f'Distribution of Market Values for {"Moroccan"} National Team Players')
plt.xlabel('Market Value (EUR)')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
sns.histplot(moroccan_players['WageEUR'], bins=20, kde=True, color='salmon')
plt.title(f'Distribution of Wages for {"Moroccan"} National Team Players')
plt.xlabel('Wage (EUR)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# basic statistics for player values and wages
value_wage_statistics = moroccan_players[['ValueEUR', 'WageEUR']].describe()

print("\nBasic Statistics for Player Values and Wages:")
print(value_wage_statistics)

```

# Player Ratings Analysis

```python
# Player Ratings Analysis for Moroccan National Team
plt.figure(figsize=(16, 8))
plt.subplot(2, 2, 1)
sns.histplot(moroccan_players['Overall'], bins=20, kde=True, color='skyblue')
plt.title(f'Distribution of Overall Ratings for {"Moroccan"} National Team Players')
plt.xlabel('Overall Rating')
plt.ylabel('Frequency')

plt.subplot(2, 2, 2)
sns.histplot(moroccan_players['Potential'], bins=20, kde=True, color='salmon')
plt.title(f'Distribution of Potential Ratings for {"Moroccan"} National Team Players')
plt.xlabel('Potential Rating')
plt.ylabel('Frequency')

plt.subplot(2, 2, 3)
sns.boxplot(x='BestPosition', y='Overall', data=moroccan_players, palette='viridis')
plt.title(f'Overall Ratings Distribution by Position for {"Moroccan"} National Team Players')
plt.xlabel('Best Position')
plt.ylabel('Overall Rating')

plt.subplot(2, 2, 4)
sns.boxplot(x='BestPosition', y='Potential', data=moroccan_players, palette='muted')
plt.title(f'Potential Ratings Distribution by Position for {"Moroccan"} National Team Players')
plt.xlabel('Best Position')
plt.ylabel('Potential Rating')

plt.tight_layout()
plt.show()
```
